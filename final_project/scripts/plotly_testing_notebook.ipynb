```python
# Import Modules
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import cartopy.feature as cfeature
import cartopy.crs as ccrs
import pandas as pd
from PIL import Image
import numpy as np
import urllib.request as ureq
import pygrib
import scipy
import shapely
import plotly.express as px # Interactive plot maker thingy
import plotly.graph_objects as go

main_path = '/home/epalmisa/met330/final_project'
#date = input('Please input date (MM/DD/YY): ')
date = '06/21/26'
if date[-2:] != '26':
    print('This script currently only supports dates in the 2026 season.')
    exit()
if len(date) != 8:
    print('Please input date in viable format. ')
    exit()
if date[0] == '0':
    date = date[1:]

# ['Game Date', 'Day of Week', 'Local Time', 'Time - ET', 'Game', 'Away Team', 'Home Team', 'Location']
df = pd.read_csv(main_path+'/schedules/2026/Full_MLB.csv')
stadium_df = gpd.read_file(main_path+'/stadium_data/2025_Stadium_Data.csv')

home_teams = []
lon_list = []
lat_list = []

for i, row in df.iterrows():
    if df['Game Date'][i] == date:
        if '-' in row['Home Team']: # Diamondbacks are only team with a hyphen
            home_team = 'Diamondbacks'
        elif ' ' in row['Home Team']:
            home_team = row['Home Team'].replace(' ','')
        else:
            home_team = row['Home Team']
        home_teams.append(home_team)

tf_list = []

for team in stadium_df['Team']:
    if team in home_teams:
        tf_list.append(True)
    else:
        tf_list.append(False)
stadium_df['home_tf'] = tf_list

team_mask = (stadium_df['home_tf']==True)

#for team in home_teams: # Putting this in here so colors are all the same
#    team_mask = (stadium_df['Team']==team)
#    lon_list.append(stadium_df[team_mask]['Longitude'])
#    lat_list.append(stadium_df[team_mask]['Latitude'])

fig = go.FigureWidget([go.Scattergeo(
        lon = stadium_df[team_mask]['Longitude'],
        lat = stadium_df[team_mask]['Latitude'],
        text = stadium_df[team_mask]['Team'],
        mode = 'markers',
        marker_color = 'red',
        )])
fig.update_layout(title=f'Baseball Games on {date}',geo_scope='usa')
fig.update_layout(hovermode='closest')



fig.show()
fig.write_html('tes.html')
```
